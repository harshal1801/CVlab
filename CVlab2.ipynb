```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow as cv2
```

```python
img=cv.imread('/content/Dots.png')
cv2(img)
```

```python
#erode
kernel = np.ones((27,27),np.uint8)
erode=cv.erode(img,kernel,iterations=1)
cv2(erode)
```

```python
dilation = cv.dilate(erode,kernel,iterations = 1)
plt.imshow(dilation, cmap='gray')
```

```python
opening = cv.morphologyEx(dilation, cv.MORPH_OPEN, kernel)
plt.imshow(opening, cmap='gray')
```

```python
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap='gray')
```
